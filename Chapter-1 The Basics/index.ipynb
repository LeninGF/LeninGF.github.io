## Definición

A computer program is said to **learn** from experience $E$ with respect to some class of tasks $T$ and performance measure $P$, if its performance at tasks in $T$, as measured by $P$, improves with experience $E$.[@mitchell97]

## Problemas de aprendizaje: Cómo Diseñaría este programa

1. Una entidad bancaria requiere diseñar un programa que permita aceptar o rechazar la otorgación de crédito a un cliente. 

2. Netflix requiere implementar un sistema de recomendación de sus series y películas a sus diversos usuarios. 

3. Se han reunido un conjunto de pacientes de cancer y se han obtenido varias muestras y resultados de pruebas de los mismos. Podemos predecir si un nuevo paciente tendrá o no la enfermedad y en que grado?

En estos problemas de aprendizaje sucede que:
1. Existe función **desconocida** $f: \mathcal{X} \rightarrow \mathcal{Y}$, donde $\mathcal{X}$ es el **input space** y $\mathcal{Y}$ es el **output space**
2. Se dispone de un conjunto de datos (**dataset**) $\mathcal{D}$ que contiene pares de ejemplos y su etiqueta (o salida): $(\mathbf{x}_{1}, y_{1}), (\mathbf{x}_{2}, y_{2}), \dots, (\mathbf{x}_{N}, y_{N}) $
3. $y_{n} = f(\mathbf{x}_{n})$ para $n=,\dots,N$
4. Existe una funcion **aprendida** $g: \mathcal{X} \rightarrow \mathcal{Y}$ que se aproxima a $f$
5. $g \in \mathcal{H}$ Donde $\mathcal{H}$ es el conjunto de hipótesis

## La desigualdad de Hoeffding
Sea $\mu$ la probabilidad de esferas rojas en una jarra y $1 - \mu$ la de esferas verdes.

Sea $\nu$ la cantidad de esferas rojas en una muestra de tamaño $N$, entonces

$\mathbb{P}[\mid \nu-\mu \mid > \epsilon] \leq 2e^{-2\epsilon^{2}N}$ y $\epsilon >0$

Mientras más grande sea la muestra $N$, es exponencialmente improbable que $\nu$ se desvie de $\mu$ más allá de la tolerancia $\epsilon$

## Extensión de la desigualdad a multiples Hipótesis:
<center>
    
   <img src="images/hoeffdingHM.png" width=50%/>
</center>

* Error en la muestra: $E_{in}(h) = \dfrac{1}{N}\sum_{n=1}^{N}[h(\mathbf{x}_{n}) \neq f(\mathbf{x}_{n})]$ y corresponde a $\nu$
* Error fuera de la muestra: $E_{out}(h) = \mathbb{P}[h(\mathbf{x}) \neq f(\mathbf{x})]$ y corresponde $\mu$
* Sin embargo, necesitamos disminuir el error de la función $g$ aprendida no de $h$, por tanto:

$\mathbb{P}[\mid E_{in}(g)-E_{out}(g) \mid > \epsilon] \leq 2e^{-2\epsilon^{2}N}$ y $\epsilon >0$

* Como existen varias hipótesis posibles:

$\mid E_{in}(g)-E_{out}(g) \mid \Rightarrow \mid E_{in}(h_{1})-E_{out}(h_{1}) \mid \lor \mid E_{in}(h_{2})-E_{out}(h_{2}) \mid \lor \dots \mid E_{in}(h_{M})-E_{out}(h_{M}) \mid $

* Dado que: 

$A_{1} \Rightarrow A_{2} \equiv \mathbb{P}[A_{1}] < \mathbb{P}[A_{2}]$

* Dado que:

$\mathbb{P}[A_{1} \lor A_{1} \lor \dots] \leq \mathbb{P}[A_{1}]+\mathbb{P}[A_{2}]+\dots$

$\mathbb{P}[\mid E_{in}(g)-E_{out}(g) \mid > \epsilon] \leq 2Me^{-2\epsilon^{2}N}$

## Enfoque probabilístico del Aprendizaje
* Un algoritmo de aprendizaje produce una hipótesis $g$ que se aproxima a una función objetivo **desconocida** $f$.
* Si el aprendizaje es **exitoso**, entonces $E_{out}(g) \approx E_{in}(g) $
* Sin embargo, **es necesario** $E_{in}(g) \approx 0$

## ¿Qué condición asegura la desigualdad de Hoeffding?
1. La desigualdad garantiza que $E_{out}(g) \approx E_{in}(g) $
2. La desigualdad garantiza que $E_{in}(g) \approx 0 $
3. 1 y 2
4. Sólo 1

## Python: Listas y Tuplas
Una lista es una coleccion de objetos de mismo o distinto tipo. Son mutables.
Las tuplas son pares ordenados y no son mutables

```python
# Este es una lista de nombres:
estudiantes_nombres = ['jose', 'lenin', 'estaban', 'alexa', 'maria']
lista = ['python', 'c++', 34, estudiantes_nombres]
lista.append('java')
print(lista)
```

```python
# Esta es una tupla
catetos = (3,5)
print('Los catetos del triangulo son: {} y {}'.format(catetos[0], catetos[1]))
```

```python
cateto1, cateto2 = catetos
cateto1 = 10
cateto2 = 10
# nueva tupla
catetos = (cateto1, cateto2)
print('Los catetos del triangulo son: {} y {}'.format(catetos[0], catetos[1]))
```
